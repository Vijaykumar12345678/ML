```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Load the training dataset
train_data = pd.read_csv('C:/Users/Admin/Downloads/train.csv')
print(train_data.info())
print(train_data.isnull().sum())
```

```python
# Drop any rows with missing values
train_data = train_data.dropna()
```

```python
print(train_data.isnull().sum())
```

```python
train_data.head()
```

```python
# Load the test dataset
test_data = pd.read_csv('C:/Users/Admin/Downloads/test.csv')
print(test_data.isnull().sum())


```

```python

# Display the first few rows to verify the data
test_data.head()
```

```python
# For training data
X_train = train_data[['x']].values  # Features (reshaped to work with sklearn)
y_train = train_data['y'].values    # Target

# For test data
X_test = test_data[['x']].values    # Features
y_test = test_data['y'].values      # Target
```

```python
# Initialize the linear regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)
```

```python
# Use the model to make predictions on the test data
y_pred = model.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE) and R-squared (R²) score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display the results
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
# Create the plot
plt.figure(figsize=(10, 6))

# Plot the training data points
plt.scatter(train_data['x'], train_data['y'], color='blue', label='Training Data')

# Plot the test data points
plt.scatter(test_data['x'], test_data['y'], color='green', label='Testing Data')

# Plot the regression line using test dataset predictions
plt.plot(test_data['x'], y_pred, color='red', linewidth=2, label='Regression Line')

# Label the axes and add a title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```
